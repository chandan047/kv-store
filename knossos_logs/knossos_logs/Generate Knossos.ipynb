```python
results = {"r/w ratio": ["0.25", "1.25", "4.00"], 
          "read latency": [],
          "write latency": [],
          "avg latency": []}

for read_write_ratio in [10, 50, 90]:
    log_path = f"ABD/ABD_{read_write_ratio}P"
    read_latencies = []
    write_latencies = []
    timestamps = []
    logs = []
    for cid in range(5):
        clog_path = path.join(log_path, f"clientlogs_{cid}.txt")
        with open(clog_path, "r") as f:
            lines = f.readlines()
            for line in lines:
                line = line.strip()
                line = line.split('\t')
                if len(line) == 3:
                    if 'read' in line[2]:
                        read_latencies.append(float(line[0]))
                    elif 'write' in line[2]:
                        write_latencies.append(float(line[0]))
                    else:
                        print ("You f**ked up")
                    line = line[1:]
                timestamps.append(line[0])
                logs.append(line[1])
                
    sorted_ids = np.argsort(timestamps)
    logs = [logs[idx] for idx in sorted_ids]
    with open(path.join(log_path, "compiled.txt"), "w") as f:
        for log in logs:
            f.write(log + "\n")
    
    print ("avg read latency", np.mean(read_latencies))
    print ("avg write latency", np.mean(write_latencies))
    print ("avg latency", np.mean(read_latencies + write_latencies))
    
    results["read latency"].append(np.mean(read_latencies) * 1000)
    results["write latency"].append(np.mean(write_latencies) * 1000)
    results["avg latency"].append(np.mean(read_latencies + write_latencies) * 1000)
    
    
    print ()
    
results = pd.DataFrame(results)
print (results.to_latex())
```
